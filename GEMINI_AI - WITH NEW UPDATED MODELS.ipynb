# USE EMBEDDINGS

```
result = genai.embed_content(
    model="models/embedding-001",
    content="What is the meaning of dream?",
    task_type="retrieval_document",
    title="Embedding of single string")

# 1 input > 1 vector output
print(str(result['embedding'])[:50], '... TRIMMED]')
```

```
result = genai.embed_content(
    model="models/embedding-001",
    content=[
      'What is the meaning of life?',
      'How much wood would a woodchuck chuck?',
      'How does the brain work?'],
    task_type="retrieval_document",
    title="Embedding of list of strings")

# A list of inputs > A list of vectors output
for v in result['embedding']:
  print(str(v)[:50], '... TRIMMED ...')
```

```
response.candidates[0].content
```

```
result = genai.embed_content(
    model = 'models/embedding-001',
    content = chat.history)

# 1 input > 1 vector output
for i,v in enumerate(result['embedding']):
  print(str(v)[:50], '... TRIMMED...')
```

# ENCODE MESSAGE

```
import google.ai.generativelanguage as glm
```

```
model = genai.GenerativeModel('models/gemini-1.5-flash-exp-0827')
response = model.generate_content(
    glm.Content(
        parts = [
            glm.Part(text="Write a short, engaging blog post based on this picture."),
            glm.Part(
                inline_data=glm.Blob(
                    mime_type='image/jpeg',
                    data=pathlib.Path('/content/drive/MyDrive/elephant 1.jpg').read_bytes()
                )
            ),
        ],
    ),
    stream=True)
```

```
response.resolve()

to_markdown(response.text[:100] + "... [TRIMMED] ...")
```

# Multiturn conversation

```
model = genai.GenerativeModel('models/gemini-1.5-flash-8b-exp-0827')

messages = [
    {'role':'user',
     'parts': ["Briefly explain how a internet impact to young child."]}
]
response = model.generate_content(messages)

to_markdown(response.text)
```

```
messages.append({'role':'model',
                 'parts':[response.text]})

messages.append({'role':'user',
                 'parts':["Okay, how about a more detailed explanation internet impact to young child in wrong or right way?"]})

response = model.generate_content(messages)

to_markdown(response.text)
```
